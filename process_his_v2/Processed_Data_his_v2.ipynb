```python
all_records = []
stop = 0
for nationId, df_nationId in df_2.groupby('NationalCard'):
    stop += 1
    n = len(df_nationId)

    sum_loanAmount = 0
    loanAmount_arr = []
    loanTime_arr = []

    for i, row in df_nationId.iterrows():
        loanAmount_arr.append(row['LoanAmount'])
        loanTime_arr.append(row['LoanTime'])

    # Tính tổng số tiền đã vay
    sum_loanAmount = np.sum(loanAmount_arr)

    # Tính TB tiền vay
    avg_loanAmount = np.average(loanAmount_arr)

    # Tính median,Min,Max
    median_loanAmount = np.mean(loanAmount_arr)
    min_loanAmount = np.min(loanAmount_arr)
    max_loanAmount = np.max(loanAmount_arr)

    # Tính TB thời gian vay
    avg_loanTime = np.average(loanTime_arr) 

    # Tính số ngày trả sớm/muộn
    sumdenta_finishdate = 0
    # Covert to datetime
    number_loan = 0
    later_lv1 = 0
    later_lv2 = 0
    later_lv3 = 0
    later_lv4 = 0
    good_loan = 0
    for i, row in df_nationId.iterrows():
        if row['FinishAt'] is not np.NaN:
            row['FinishAt'] = row['FinishAt'].split(" ")[0]
            row['FinishAt'] = datetime.datetime.strptime(row['FinishAt'], '%Y-%m-%d')

        if row['ToDate'] is not np.NaN:
            row['ToDate'] = datetime.datetime.strptime(row['ToDate'], '%Y-%m-%d')
        
        if row['FinishAt'] is not np.NaN and row['ToDate'] is not np.NaN:
            denta_finishdate = (row['FinishAt'] - row['ToDate']).days
        else:
            denta_finishdate = 0

        
        ''' Đơn vay trả muộn
        Mức 1: 7 => 29 days
        Mức 2: 30 => 59 days
        Mức 3: 60 => 89 days
        Mức 4: >89 days
        '''
        number_loan += 1
        # Tính số lần trả muộn theo mức
        if denta_finishdate <= 7:
            good_loan += 1
        elif denta_finishdate >= 7 and denta_finishdate < 30:
            later_lv1 += 1
        elif denta_finishdate >= 30 and denta_finishdate < 60:
            later_lv2 += 1
        elif denta_finishdate >= 60 and denta_finishdate < 90:
            later_lv3 += 1
        elif denta_finishdate >= 30:
            later_lv4 += 1

    
    sumdenta_finishdate += denta_finishdate
    

    record = {
        "NationalCar" : nationId,
        "sum_LoanAmount" : sum_loanAmount,
        "avg_loanAmount" : avg_loanAmount,
        "median_loanAmount" : median_loanAmount,
        "min_loanAmount" : min_loanAmount,
        "max_loanAmount" : max_loanAmount,
        "avg_loanTime" : avg_loanTime,
        "number_loan" : number_loan,
        "good_loan" : good_loan,
        "later_lv1" : later_lv1,
        "later_lv2" : later_lv2,
        "later_lv3" : later_lv3,
        "later_lv4" : later_lv4
    }
    all_records.append(record)
    
all_records = pd.DataFrame(all_records)
# all_records.to_csv('data_his_v2_processed.csv')
```

```python
result_all = []
for nationId, df_nationId in df.groupby('NationalCard'):
    for i, row in df_nationId.iterrows():
        if row['CreatedTime'] is not np.NaN:
            row['CreatedTime'] = row['CreatedTime'].split(" ")[0]
            date = datetime.datetime.strptime(row['CreatedTime'], '%Y-%m-%d')
            # Thứ trong tuần!
            day_of_week_test = date.strftime('%A')
            if day_of_week_test == 'Monday':
                day_of_week = 2
            elif day_of_week_test == 'Tuesday':
                day_of_week = 3
            elif day_of_week_test == 'Wednesday':
                day_of_week = 4
            elif day_of_week_test == 'Thursday':
                day_of_week = 5
            elif day_of_week_test == 'Friday':
                day_of_week = 6
            elif day_of_week_test == 'Saturday':
                day_of_week = 7
            elif day_of_week_test == 'Sunday':
                day_of_week = 1
            else:
                day_of_week = -1
            # GDP theo tháng
            '''
            IV, 2020: 4.48
            I, 2021: 4.72
            II, 2021: 6.61
            III, 2021: -6.02
            IV, 2021: 5.22
            I, 2022: 5.03
            '''
            # CPI theo tháng
            '''
            IV, 2020: 1.38
            I, 2021: 1.92
            II, 2021: 0.45
            III, 2021: 3.72
            IV, 2021: 1.89
            I, 2022: 1.92
            '''
            err = []
            if date >= datetime.datetime(2021, 1, 1) and date < datetime.datetime(2021, 4, 1):
                gdp = 4.48
                cpi = 1.38
            elif  date >= datetime.datetime(2021, 4, 1) and date < datetime.datetime(2021, 7, 1):
                gdp = 4.72
                cpi = 1.92
            elif  date >= datetime.datetime(2021, 7, 1) and date < datetime.datetime(2021, 10, 1):
                gdp = 6.61
                cpi = 0.45
            elif  date >= datetime.datetime(2021, 10, 1) and date < datetime.datetime(2022, 1, 1):
                gdp = -6.02
                cpi = 3.72
            elif  date >= datetime.datetime(2022, 1, 1) and date < datetime.datetime(2022, 4, 1):
                gdp = 5.22
                cpi = 1.89
            elif  date >= datetime.datetime(2022, 4, 1) and date < datetime.datetime(2021, 7, 1):
                gdp = 5.03
                cpi =1.92
            else:
                print(date)
                err.append(date)
                gpd = -1
                cpi = -1

        result = {
            "LoanBriefId": row['LoanBriefId'],
            "NationalCar" : nationId,
            "CreatedTime": row['CreatedTime'],
            "day_of_week": day_of_week,
            "gdp": gdp,
            "cpi": cpi
        }
    result_all.append(result)
result_all = pd.DataFrame(result_all)

result_all
```
